```python
## import all requisite modules ##
from continuous_annealing import Simm_Anneal
import numpy as np
import matplotlib.pyplot as plt
```

# Define the cost function to send into the optimization algorithm

```python
def f(x):
    """ Function to minimize."""
    return 0.2*x**2 + np.sin(10*x)

def cost_function(x):
    """ Cost of x = f(x)."""
    return f(x)
```

# Define the parameters to use and run optimization process

```python
maxsteps = 3000
debug = False
interval =(-10,10)

# Run the simulated annealing optimization process
state, cost_functions, states, costs = Simm_Anneal(maxsteps, cost_function, interval, debug).anneal()
```

```python
# Visualize the actual function to minimize
x = np.linspace(interval[0], interval[1], 1000)
plt.figure(figsize=(10,7))
plt.plot(x, f(x))
plt.xlim(interval)
plt.grid()
plt.title('Actual distribution for provided f(x)')
plt.show()

# Visualize the predicted minimal position over iterations
plt.figure(figsize=(10,7))
plt.scatter(range(1,len(states)+1), states)
plt.grid()
plt.xlabel('Iteration')
plt.ylabel('State location (x)')
plt.title('State minima over time (actual for x just less than 0)')
plt.show()

print('As can be seen, this does converge to the reasonably correct area')
```
