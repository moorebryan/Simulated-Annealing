```python
from traveling_salesman import Simm_Anneal
```

```python
maxsteps=5000
debug=False

coords = []
with open("coords.txt", "r") as f:
    for line in f.readlines():
        line = [float(x.replace("\n", "")) for x in line.split(" ")]
        coords.append(line)

if __name__ == "__main__":
    sa = Simm_Anneal(coords, maxsteps, debug)
    sa.anneal()
    sa.visualize_routes()
    sa.plot_learning()
    
```
